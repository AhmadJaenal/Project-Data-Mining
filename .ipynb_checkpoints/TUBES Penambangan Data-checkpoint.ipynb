# Pemanggilan library yang digunakan dan juga pengambilan dataset

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
pd.set_option("display.max_rows", df.shape[0] + 1)

df.head(10)
```

# Selection

```python
daftar_fitur_yang_tidak_digunakan = ['Person ID']
df = df.drop(daftar_fitur_yang_tidak_digunakan, axis = 1)
df.head()
```

```python
# Jumlah data awal yang belum diproses
grouped = df.groupby('Sleep Disorder').count()

display(grouped.iloc[:, 0:1])
```

# Data Preprocessing

```python
# Pengecekan Missing Value
df = df.dropna(how='all')
data_null = df.isnull().sum()
display(data_null)
```

```python
# Penghapusan data duplikat
df = df.drop_duplicates()

# Data disimpan difile yang berbeda agar dapat terlihat perbedaanya  
df.to_csv('data_preprocessing.csv', index=False)
```

```python
df = pd.read_csv('data_preprocessing.csv')
df.shape
```

# Proses binerisasi dan normalisasi pada data

```python
# Fitur Gender
df['Gender'] = df['Gender'].replace('Male', 1)
df['Gender'] = df['Gender'].replace('Female', 0)

# Normalisasi fitur daily steps
preprocessor = preprocessing.MinMaxScaler(feature_range=(0,1))
scaled_data = preprocessor.fit_transform(df[['Daily Steps']])

for i in range(len(scaled_data)):
    df.loc[i, 'Daily Steps'] = scaled_data[i][0]

binerisasi_BMI = {'Normal': 0, 'Normal Weight': 0, 'Overweight': 1, 'Obese' : 2}
df['BMI Category'] = df['BMI Category'].map(binerisasi_BMI)

binerisasi_sleep_disorder = {'None': 0, 'Insomnia': 1, 'Sleep Apnea': 2}
df['Sleep Disorder'] = df['Sleep Disorder'].map(binerisasi_sleep_disorder)

binerisasi_occupation = {'Software Engineer' : 0, 'Doctor' : 1, 'Sales Representative' : 2, 'Teacher' : 3,
       'Nurse' : 4 , 'Engineer' : 5, 'Accountant' : 6, 'Scientist' : 7, 'Lawyer' : 8,
       'Salesperson' : 9, 'Manager' : 10}
df['Occupation'] = df['Occupation'].map(binerisasi_occupation)

for i in range(len(df)):
    blood_pressure = df['Blood Pressure'][i].split('/')
    df.loc[i, 'Systolic'] = int(blood_pressure[0])
    df.loc[i, 'Diastolic'] = int(blood_pressure[1])

new_columns = ['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'BMI Category', 'Systolic', 'Diastolic', 'Heart Rate', 'Daily Steps', 'Sleep Disorder']
    
data = pd.DataFrame(df, columns=new_columns)

data.head()
```

# Penerapan Algoritma KNN

```python
# Variable independen
X = data.drop(['Sleep Disorder'], axis = 1)
X.head()

# Variable dependen
y = data['Sleep Disorder']
y.head()
```

```python
# Proses pembagian data menjadi data train dan test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.21, random_state=458)
```

```python
# Proses dengan algoritma KNN dengan nilai k = 5
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# Data untuk test
gender = 1
age = 29
occupation = 3
sleep_duration = 6.3
quality_of_sleep = 6
physical_activity_level = 40
stress_level = 7
bmi_category = 2
systolic = 140
diastolic = 90
heart_rate = 82
daily_steps = 0.157423	

data.head(1)

X_new = np.array([gender, age, occupation, sleep_duration, quality_of_sleep, physical_activity_level, stress_level, bmi_category, systolic, diastolic, heart_rate, daily_steps])
X_train = np.array(data.drop(columns="Sleep Disorder"))
X_train
```

```python
# Proses penghitungan nilai distance
from scipy.spatial.distance import euclidean

distance = [euclidean(X_new, d) for d in X_train]
distance
```

```python
# Pengurutan nilai distance
data['Distance'] = distance
data.sort_values(['Distance'])
data
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# Pengecekan nilai classification report
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)
print(cls_report)
```
